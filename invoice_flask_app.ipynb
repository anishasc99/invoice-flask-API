```python
from flask import Flask, flash, request, redirect, url_for
from flask_ngrok import run_with_ngrok
from flask_wtf import FlaskForm
from wtforms import RadioField, StringField, PasswordField, BooleanField, SubmitField
from flask_wtf.file import FileField, FileRequired
from flask import Flask, render_template
from wtforms.validators import DataRequired, InputRequired
from werkzeug.utils import secure_filename
import subprocess
import shutil
from shutil import copyfile
import os
import time
import json
import cv2
import pytesseract
from pytesseract import Output
import re

app = Flask(__name__)
app.config['SECRET_KEY'] = 'filesystem111'

UPLOAD_FOLDER = '.'
ALLOWED_EXTENSIONS = { 'pdf', 'png', 'jpg', 'jpeg'}
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER

run_with_ngrok(app)  

class InvoiceForm(FlaskForm):
  invoice_file = FileField("file")
  #algorithm = RadioField('algorithm', choices=[(0,'SSD'),(1,'YOLO'), (2,'Faster R-CNN')], coerce=int, validators=[InputRequired()])
  submit = SubmitField('Parse')

  

def allowed_file(filename):
    return '.' in filename and \
           filename.rsplit('.', 1)[1].lower() in ALLOWED_EXTENSIONS

def subprocess_cmd(command):
    process = subprocess.Popen(command,stdout=subprocess.PIPE, shell=True)
    proc_stdout = process.communicate()[0].strip()
    for line in proc_stdout.decode().split('\n'):
        print (line)

def clean(folder):
  for filename in os.listdir(folder):
      file_path = os.path.join(folder, filename)
      try:
          if os.path.isfile(file_path) or os.path.islink(file_path):
              os.unlink(file_path)
          elif os.path.isdir(file_path):
              shutil.rmtree(file_path)
      except Exception as e:
          print('Failed to delete %s. Reason: %s' % (file_path, e))


def ocr(dir, output):
  arr = os.listdir(dir)
  custom_oem_psm_config = r'--oem 1'
  outputdict = {}

  for f in arr:
    key = f[10:-4]
    f = dir + "/"+f
    crop_img = cv2.imread(f)
    d = pytesseract.image_to_string(crop_img, output_type=Output.DICT,config = custom_oem_psm_config)
    d = re.sub('\n', ' ', d['text'])
    d = re.sub(r'[\x00-\x08\x0b\x0c\x0e-\x1f\x7f-\xff]', '', d)
    outputdict[key] = d
    print(key,d)
  with open(output, 'w') as outfile:
          json.dump(outputdict, outfile)

@app.route("/", methods=['GET', 'POST'])



def invoice():
  form = InvoiceForm()
  algos = ['SSD', 'YOLO', 'Faster R-CNN']
  #algorithm = -1
  filename = ''
  if request.method == 'POST':
        if 'invoice_file' not in request.files:
            print("No file part")
            return redirect(request.url)
        file = request.files['invoice_file']
        if file.filename == '':
            print('No selected file')
            return redirect(request.url)
        if file and allowed_file(file.filename):
            filename = secure_filename(file.filename)
            file.save(os.path.join(app.config['UPLOAD_FOLDER'], filename))
        #algorithm = request.form['algorithm']
        algo = request.form['algos']
  
  if form.validate_on_submit():
    result_img = './static/styles/'
    entities = {}
    image = filename
    #if algorithm == '0':
    if algo == "SSD":
      model = "weights/COCO (1).pth"
      command = 'cd ssd-custom-dataset; python test_invoice.py --trained_model "'+model+'" --image ' +image
      subprocess_cmd(command)
      result_img = result_img+'test_img_ssd.jpg'
      shutil.copyfile('./static/test_img_ssd.jpg', result_img) 
      f = open('./ssd-custom-dataset/output.json',)
      entities = json.load(f)
    #if algorithm == '1':
    if algo == "YOLO":
      model = "yolov4-custom_3000.weights"
      image = "/content/invoice-flask-API/"+filename
      folder = './darknet/result_img'
      output = 'output_yolo.json'
      clean(folder)      #clean result folder (with cropped images) before testing new image
      command = "cd darknet; ./darknet detector test data/yolo.data cfg/yolov4-custom.cfg '"+model+"' '"+image+"' -thresh 0.40"
      subprocess_cmd(command)
      src = "./darknet/predictions.jpg"
      result_img = result_img+'test_img_yolo.jpg'
      shutil.copyfile(src, result_img) 
      ocr(folder, output)  #perform ocr on cropped images
      f = open('output_yolo.json')
      entities = json.load(f)
    #if algorithm == '2':
    if algo == "Faster R-CNN":
      folder = "./detectron2-custom/cropped images"
      command = 'cd detectron2-custom; python test_one.py'
      test_folder = "./detectron2-custom/test"
      output = "output_rcnn.json"
      clean(folder)
      clean(test_folder)
      shutil.copyfile(image, test_folder+"/"+image) 
      subprocess_cmd(command)
      src = "./detectron2-custom/predictions_detectron.jpg"
      result_img = result_img+'test_img_rcnn.jpg'
      shutil.copyfile(src, result_img) 
      ocr(folder, output)
      f = open('output_rcnn.json')
      entities = json.load(f)

     
    print(result_img)
    print(entities)
    return render_template("result.html", result_img = result_img, entities = entities)
  else:
    print (form.errors)
  return  render_template("form.html", algos=algos, form=form)
    
if __name__ == '__main__':
   app.run()
```
